## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
sns.barplot(x="PURPOSE*",y="MILES*",data=uber_drives);
plt.xticks(rotation=90)
plt.show()
```

```python
# This Graph gives us information about the distance travelled(In Miles) by uber drives for various purposes.i.e. uber drives travelled maximum for commute purpose ,then for customer visit then for meeting & least for errand/supplies.This Graph shows the clear information of purpose VS distance travelled (In Miles). 
```

## Q20. Display a dataframe of Purpose and the total distance travelled for that particular Purpose. (3 points)
#### Note: Use the original dataframe without dropping "NA" values

```python
data_frame_purpose_total_distance = uber_drives.groupby(["PURPOSE*"])["MILES*"].sum()
data_frame_purpose_total_distance
data_frame_purpose_total_distance.reset_index()
```
